# Trocando nomes das colunas

```python
# Pedidos
df_pedidos_raw.columns = ['id', 'data', 'hora', 'id_membro', 'id_restaurante', 'valor_pedido']
df_pedidos_raw.head()
```

```python
# Restaurantes
df_restaurantes_raw.columns = ['id', 'nome_restaurante', 'id_tp_restaurante', 'renda_percentual', 'id_cidade']
df_restaurantes_raw.head()
```

```python
# Tipos de Refeição
df_tp_refeicao_raw.columns = ['id', 'tp_refeicao']
df_tp_refeicao_raw.head()
```

```python
# Tipos de Restaurante
df_tp_restaurante_raw.columns = ['id', 'tp_restaurante']
df_tp_restaurante_raw.head()
```

```python
# Tipos de Serviço
df_tp_servico_raw.columns = ['id', 'tp_servico']
df_tp_servico_raw.head()
```

```python
# Total Mensal de Membros
df_ttl_mens_membros_raw.columns = ['id_membro', 'nome', 'sobrenome', 'genero', 'email', 'cidade', 'ano', 'mes', 
                                   'qtd_pedidos', 'qtd_refeicoes', 'orcamento_mensal', 'despesa_total', 'saldo', 'comissao']
df_ttl_mens_membros_raw.head()
```
